```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

$$ \dot{y} = f(t, y) $$

### we will use the differential equation $y'(t) = y(t)$.  The analytic solution is $y = e^t$.


```python

def f(t,y):
    return y

def asol(t):
    return math.exp(t)

yasol = np.vectorize(asol)
```

```python
dt = 0.1
t0 = 0.0
y0 = 1.0 # initial condition

t = np.arange(0.0, 10.0, dt)
y = np.zeros(t.size)
ye = np.zeros(t.size)
y[0] = y0
ye[0] = y0
```

```python
for i in range(1, t.size):
    y_intermediate = y[i-1] + dt*f(t[i-1],y[i-1])

    y[i] = y[i-1] + (dt/2.0)*(f(t[i-1],y[i-1]) + f(t[i],y_intermediate))
    ye[i] = y_intermediate
```

```python
plt.plot(t, np.abs(y-yasol(t))/yasol(t), '.', label="RK2")
plt.plot(t, np.abs(ye-yasol(t))/yasol(t), '.', label="Euler")
plt.legend()
plt.show()
```

```python
plt.plot(t, y, '.', label="RK2")
plt.plot(t, ye, '.', label="Euler")
plt.plot(t, yasol(t), '-', label="Analytic")
plt.legend()
plt.show()
```
